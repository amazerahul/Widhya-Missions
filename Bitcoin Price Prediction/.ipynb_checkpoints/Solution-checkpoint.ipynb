### What funtions in pandas allows you to fill cells in a column that have NaN values? Just mention the name of the function.

Ans: fillna

```python
columns_filtered = ['btc_market_cap', 'btc_n_transactions', 'btc_miners_revenue', 'btc_cost_per_transaction', 'btc_difficulty', 'btc_hash_rate', 'btc_cost_per_transaction_percent', 'btc_market_price']
```

```python
data1 = data[columns_filtered]
```

```python
data1.isnull().sum()
```

```python
data1['btc_difficulty'].fillna(data1['btc_difficulty'].mean(), inplace=True)
```

### What is the MSE of your model? Round off to the nearest integral value.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = data1.drop('btc_market_price', axis=1)
y = data1['btc_market_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
round(mean_squared_error(y_test, y_pred))
```
